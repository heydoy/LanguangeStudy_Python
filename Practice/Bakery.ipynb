```

# 저희 다말베이커리에 오신것을 환영합니다!!!
# 빵셔틀

import ast,sys      # 가져다쓸 라이브러리 (설치방법 : cmd창에서 pip install ast, 아나콘다 설치했으면 이미 있을듯)


bakery = []     # 빵집DB 관리용 리스트 선언, 전역 변수, 다른 함수에서도 쓸수있게 (참고로 함수내에서만 쓰이는 변수 = 지역변수)
client = []     # 고객DB 관리용 리스트 선언, 전역 변수, 다른 함수에서도 쓸수있게 (전역변수는 함수 바깥에서 선언이 됨)
path = "C:\\class\\"            # (로컬(오프라인)에서 구동할때만) 자기 컴퓨터 C드라이브에 class 폴더 만들어서 python파일 저장해주세요

def load_info(name):
    global bakery, client,money       # global 선언은 해당 전역변수를 함수내에서도 가져다 쓰겠다는말 (함수내에서 수정도 가능)
     
    try :
        content = []
        f = open(path+name+".txt",'r')      # 읽기전용 모드로 파일 열기 
        content_str = f.read()              # 파일 내용 읽어서 content_str 이란 변수에 몽땅 '문자열' 형식으로 넣기
        if content_str and name != "money":                  # 파일 내용이 비어있지 않다면 
            content = ast.literal_eval(content_str)  # 문자열을 리스트로 변환해주는 라이브러리
        f.close()
        if name == "money" and content_str != '' :  money = int(content_str)
        elif name == "money" and content_str == '' :  money = 0       # 빵집 수입 
        if name == "bakery" and content_str != '' : bakery = content     # 미리 할당해논 전역변수에 파일내용 담기 
        elif name == "bakery" and content_str == ''  :  # 처음 실행이라면 
            register_bread("소보루",40,1000)    # 빵 추가하기 
            register_bread("피자빵",10,1500)    # 빵 추가하기 
            register_bread("초코빵",25,2000)    # 빵 추가하기
            register_bread("크림모카빵",77,2500)    # 빵 추가하기
            register_bread("호두식빵",33,3000)    # 빵 추가하기
        if name == "client"  : client = content
    except FileNotFoundError :   # 파일이 없다면 
        f = open(path+name+".txt",'w')  # 새로 생성하기
        if name == "bakery"  :
            register_bread("소보루",40,1000)    # 빵 추가하기 
            register_bread("피자빵",10,1500)    # 빵 추가하기 
            register_bread("초코빵",25,2000)    # 빵 추가하기
            register_bread("크림모카빵",77,2500)    # 빵 추가하기
            register_bread("호두식빵",33,3000)    # 빵 추가하기
        if name == "money" :
            money = 0
        f.close()
    
		
def write(name): # 정보 입력 
    global bakery, client
    f = open(path+name+".txt",'w')      # 전부 다시 쓰기 모드로 파일 열기 
    if name == "bakery" :
        f.write("[\n")              # 일부러 2차원 배열로 만들어서 저장하기 위해 대괄호 추가
        for i in range(0,len(bakery)):     # 가독성 있게 한줄한줄 개행해서 저장 
            f.write(str(bakery[i])+",\n")   # 그래서 문자열 형태로 저장 
        f.write("]")  
    if name == "client" :
        f.write("[\n")
        for i in range(0,len(client)):     # 가독성 있게 한줄한줄 개행해서 저장 
            f.write(str(client[i])+",\n")  #  \n는 한줄 밑으로 가는 개행문자
        f.write("]")   
   
    f.close()

def is_first(name) :  # 처음 방문인지 여부에 따라 , 민진님 
    global bakery, client
    phrase = ""
    for i in range(0,len(client)):
        if name == client[i][0] :      # 고객정보 이름과 같다면 
            phrase = "또"*client[i][1]+" 오셨네요! "+name+"님"     # 멘트 준비
            update_client(name,1)           # 고객 방문횟수 업뎃
            return phrase                   # 멘트 반환하며 함수 종료 

    phrase = name+"님,"+" 첫 방문 감사합니다."          # 위에서 반환되지않고 여기까지 왔다면 일치하는 고객이름 못찾은 것이므로
    update_client(name,0)               # 고객DB에 고객정보 새로 추가하고 
    return phrase                   # 리턴값이란 함수 종료시키며 특정값을 반환(output)함 (함수마다 리턴을 꼭 해야하는건 아님)


def register_bread(bread,amount,price):  # 운영자가 쉽게 새로운 빵을 DB에 추가하기위함 
    # 빵이름, 재고수량, 가격,
    bakery.append([bread,amount,price])  # 전역변수에 새로운 빵 종류 추가 
    write("bakery")             # 전역변수 내용을 파일에 저장하기
    
def del_bread(index):  # 운영자가 쉽게 등록된 빵을 DB에서 삭제하기위함 
    # 빵이름, 재고수량, 가격, 
    bakery.pop(index)  # 전역변수에 새로운 빵 종류 추가 
    write("bakery")             # 전역변수 내용을 파일에 저장하기
    #load_info("bakery")  


def update_client(name,op):  # 0:없으면 추가, 1:있으면 방문횟수 증가 
    global bakery, client
    if op == 0 :            # 고객 DB에 새로 추가 
      
        client.append([name,1])     # 구매할 빵과 수량 정보도 추가적으로 넣을것임
        for i in range(len(bakery)):
            client[-1].append([bakery[i][0],0])
        #print(client)       # 확인용
        write("client")         # 파일에 쓰기 
    if op == 1 :            # 고객 DB에 기존에 있어서 방문횟수만 수정 
        for i in range(0,len(client)):    # 고객정보 하나하나 검토하면서
            if client[i][0] == name :     # 고객이름과 매치되는 이름 찾으면 
                client[i][1] = int(client[i][1]) + 1   # 방문횟수 1 늘리기 
        write("client")       # 방문횟수 변동되었으니 새로 파일 쓰기 

def welcome():  # 처음 고객 응대 
    name = input("고객님 성함을 입력해주세요:")
    print(is_first(name))       # 첫 방문인지 여부 확인해 멘트 날리고, 고객정보업뎃
    return name
  
def check_input(co):   # 여러입력을 숫자로 통일시켜 output , 재형님 
    if "개" in co:
        co = co.replace('개','')
    if "번" in co:
        co = co.replace('번','')
    return int(co)
  

def check_sold_out():     # 민진님 
    cnt = 0
    for i in range(len(bakery)) : 
        if bakery[i][1] == 0:
            cnt = cnt+1
    if cnt == 5:
        return 1
    else:
        return 0
  
def transaction(name):   # 거래관리, 메인함수, 민진님 
    global money
    
    while True:
        showMenu()
        while True:
            if check_sold_out() == 1 : 
                print("모든 빵 완판되었습니다. 감사합니다.")
                save_income(money)
                sys.exit()     # 0번 누를시 프로그램 종료 
            if client_most() != -1 :
                print(name+"님이 젤 좋아하시는 "+client_most()[0]+" 드릴까요?")
            sel = input("\n어떤빵 구매하시겠습니까? (번호입력, 0번은 매장 나가기):")
            
            if sel == "0" : 
                save_income(money)
                sys.exit()     # 0번 누를시 프로그램 종료 
            if check_input(sel) > len(bakery):
                print("제대로 입력해주세요.")
            else: break
            
        sel = check_input(sel)
        bread = bakery[sel-1][0]
        print(">>>>{} 고르셨네요.".format(bread,end="."))
        
        while True :
          amount = input("몇개 구매하시겠습니까? 현재 {}개 있습니다:".format(bakery[sel-1][1]))
          amount = check_input(amount)
          if amount > 0 : break
          else : print("제대로 입력해주세요.")
        
        
        for i in range(len(bakery)):
            if bakery[i][0] == bread:
                if bakery[i][1] == 0:
                    print("아쉽지만 재고가 다 떨어졌네요.")
                    continue
                    
                elif bakery[i][1]-amount < 0 : 
                    print("그 빵은 원하시는만큼 있지가 않네요.")
                    continue
            
                else:
                    
                    print("{}개 드릴게요. 맛있게 드세요!".format(amount))
                    money += bakery[sel-1][2] * amount     # 수입만큼 자산에 추가
                    save_income(money)                     # 자산 업데이트  
                    print("현재까지 빵집 매출:",money,"원")
                    stock_manager(bread,amount)  # 재형님
                    update_client_transac(name,bread,amount)   # 도이님 
                    
                    
def save_income(money):
    f = open(path+"money.txt",'w')   # 종료 선택시 수입 저장
    f.write(str(money))
    f.close()
    
            
def showMenu():     # 도이님
    print("_"*36)
    print("."*36);print("."*12+" 다말 베이커리  "+"."*12,end='\n');print("."*12+"    메뉴판    "+"."*12,end='\n')
    print("."*36,end='\n\n')
    for i in range(len(bakery)):
        print("          {}. {} : {}원".format(i+1,bakery[i][0],bakery[i][2]))

    print("_"*36)
    print("")
    
def stock_manager(bread,amount) :      # 재고 증감시키는 함수 , 재형님 
    global bakery
    for i in range(len(bakery)):
      if bakery[i][0] == bread:
        bakery[i][1] = bakery[i][1]-amount
    #print(bakery)                # 확인용
    write("bakery")

def update_client_transac(name,bread,amount):   # 고객의 구매물품과 수량 정보 업뎃하기, 도이님
    num = 0
    if bread == "소보루": num = 1
    if bread == "피자빵": num = 2
    if bread == "초코빵": num = 3
    if bread == "크림모카빵": num = 4
    if bread == "호두식빵": num = 5
      
    for i in range(len(client)):
        if client[i][0] == name :       # 고객 이름 일치하는것 찾아서 
            client[i][num+1][1] +=  client[i][num+1][1] +amount    # 고객이 구매한 빵 수량 증가시키기 

def client_most():  # 젤 마니 구매한 빵 구하기 , 도이님 
    most = 0
    index = 0 
    for i in range(len(client)):    # 고객 DB에서 
        for k in range(5):           # 고객이 구매했던 빵 종류별로 훑어가며
            
            if int(client[i][k+2][1]) > most :    # 제일 구매 수량 많은것 찾기 
                most = client[i][k+2][1]
                index = k+2
    
    if most == 0 :
        return -1
    else : 
        return client[i][index]
            



##### 본문 #####   

##### 처음 구동시에만 실행하고 말것들
load_info("money")      # 빵집 수입 가져오기 
load_info("bakery")     # 빵집 정보 가져오기
load_info("client")     # 고객 정보 가져오기 
name = welcome()      # 첫 응대 , 방문 횟수 분석해서 멘트 날리는함수 호출 


#del_bread(2)     # 인덱스 번호로 빵 지우기 


##### 이하 반복 작업 
while True :     
    transaction(name)     # 구매 질문-답변 반복 
    

"""
데이터 형식 : bakery = [[빵1, 재고, 가격],[빵2,재고,가격],[],[],[]....]
데이터 형식 : client = [[홍길동,방문횟수,[[구매물품1,구매수량],[구매물품2,구매수량],[구매물품3,구매수량]],
                        [철수,방문횟수,[[],[],[]]
                        [이름,방문횟수,[[],[]]
                        ....]
                        
                        [홍길동,방문횟수[[빵,수량][빵,수량]]]
# 구현이 필요한 것들 
# -  빵종류 추가, 멘트 다듬기 , transaction(거래) 함수 ,  재고증감 함수 , 수입 데이터, 자산관리 함수 등 """
```
